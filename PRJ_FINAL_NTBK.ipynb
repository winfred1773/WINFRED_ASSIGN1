##data cleaning

For the intake and outcome files, I make them easier to work with by normalizing column names,getting the annual production and dropping unnecessary columns

```python
#to get the total number of movies in each year
df['year'].value_counts()
```

```python
#to get for a specific year -2018
df[df['year']==2018]
```

```python
#to find out which foreign_gross recorded a NaN
df[df['foreign_gross'].isnull()]
```

```python
#sort title names alphabetically
df.sort_values('title')
```

```python
#find the movies with the highest record
df.sort_values('domestic_gross',ascending=False)
```

```python
#to get the percentage of foreign_gross
df['foreign_gross'].value_counts(normalize=True)
```

```python
#to get the percentage of domestic_gross
df['domestic_gross'].value_counts(normalize=True)
```

```python
#to get the percentage difference in domestic and foreign_gross
df['foreign_gross'].value_counts(normalize=True)-df['domestic_gross'].value_counts(normalize=True)
```

```python
#to get the missing value
df.isnull().sum()
```

```python
#to get which year had the highest record
df['year'].value_counts()
```

```python
#to get which studio had the highest number of movies
df['studio'].value_counts()
```

```python
#to draw a bar graph
df['year'].value_counts().plot(kind='bar')
```

```python
#to draw a box plot
df['year'].plot(kind='box')
```

```python
#to draw a line graph
df['year'].value_counts().plot(kind='line')
```

```python
#to draw a scatter graph
df.plot(kind='scatter',x='year',y='domestic_gross')
```
